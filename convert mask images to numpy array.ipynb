```python
from matplotlib import pyplot as plt, cm
import numpy as np
import glob, os
import h5py
import ntpath
from PIL import Image



DATA_DIR_TRAINING = "C:/Users/Kevin/Desktop/patient data/mask/train"
DATA_DIR_TESTING = "C:/Users/Kevin/Desktop/patient data/mask/test"

patients_training_mask = glob.glob(os.path.join(DATA_DIR_TRAINING, '*')) # Get the folder names for the patients - for training
patients_testing_mask = glob.glob(os.path.join(DATA_DIR_TESTING, '*'))

training_mask = []
testing_mask = []

#for training
for patientDirectory in patients_training_mask[:1]:
    image = Image.open(patientDirectory).convert('L')
    img_arr = np.array(image)
    training_mask = np.expand_dims(img_arr, axis = 0)
    
    
for patientDirectory in patients_training_mask[1:]:
    image = Image.open(patientDirectory).convert('L')
    img_arr = np.array(image)
    training_mask = np.append(training_mask, np.expand_dims(img_arr, axis = 0), axis = 0)
print(training_mask.shape)
    
            
#for testing

for patientDirectory in patients_testing_mask[:1]:
    image = Image.open(patientDirectory).convert('L')
    img_arr = np.array(image)
    testing_mask = np.expand_dims(img_arr, axis = 0)
    
    
for patientDirectory in patients_testing_mask[1:]:
    image = Image.open(patientDirectory).convert('L')
    img_arr = np.array(image)
    testing_mask = np.append(testing_mask, np.expand_dims(img_arr, axis = 0), axis = 0)
print(testing_mask.shape)

for img in training_mask[:1]:
    print(img)
    plt.imshow(img)
    plt.show()
```
