```python

# Importing Numpy package
import numpy as np
import random
from matplotlib import pyplot as plt, cm

# Initializing value of x-axis and y-axis
# in the range -1 to 1
x, y = np.meshgrid(np.linspace(-1,1,512), np.linspace(-1,1,512))
dst = np.sqrt(x*x+y*y)
  
# Intializing sigma and muu
sigma =random.uniform(0.3, 0.5)
muu   =random.uniform(2, 2.1)
  
# Calculating Gaussian array
gauss = np.exp(-( (dst-muu)**2 / ( 2.0 * sigma**2 ) ) )

for array in gauss:
    indices = np.random.choice(np.arange(array.size), replace=False,size=int(array.size * 0.3))
    array[indices] = 0
print("2D Gaussian array :\n")

plt.subplot(121)
plt.imshow(gauss, cmap='gray')
plt.show()
plt.axis('off')
plt.title('Mask')
plt.subplot(122)
plt.imshow(np.fft.fftshift(gauss))
plt.show()
plt.axis('off')
plt.title('FFT')
```
